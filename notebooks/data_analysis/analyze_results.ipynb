### 0. Import libraries and import data

```python
import math
import numpy as np
import matplotlib
import seaborn as sns
from matplotlib import pyplot as plt

sns.set()
matplotlib.rcParams.update({'font.size': 24})
```

### 1. Plot confidence intervals

```python
def plot_confidence_interval(x, values, z=1.96, color='#7848aa', horizontal_line_width=0.25):
    marker = 'o'

    mean = np.array(values).mean() / 100
    stdev = np.array(values).std() / 100
    confidence_interval = z * stdev / math.sqrt(len(values))

    left = x - horizontal_line_width / 2
    top = mean - confidence_interval
    right = x + horizontal_line_width / 2
    bottom = mean + confidence_interval
    print(f"{round(mean, 4)} +- {round(confidence_interval, 4)}")

    plt.plot([x, x], [top, bottom], color=color, linewidth=4)
    plt.plot([left, right], [top, top], color=color, linewidth=4)
    plt.plot([left, right], [bottom, bottom], color=color, linewidth=4)
    plt.plot(x, mean, marker, color='#6ec6ca', markersize=12)
```

```python
def plot_save_conf_intervals(optimized, pos_after, pos_between, ner_span_multi_4, ner_span_multi_18, ner_span_single_4, ner_span_single_18, metric):
    fig, _ = plt.subplots(1, 1)
    fig.set_size_inches(14, 5)

    plot_confidence_interval(1, optimized)
    plot_confidence_interval(2, pos_after)
    plot_confidence_interval(3, pos_between)
    plot_confidence_interval(4, ner_span_multi_4)
    plot_confidence_interval(5, ner_span_multi_18)
    plot_confidence_interval(6, ner_span_single_4)
    plot_confidence_interval(7, ner_span_single_18)

    plt.xticks([1, 2, 3, 4, 5, 6, 7], ['Optimized', 'POS after', 'POS between', 'NE multi 4', 'NE multi 18', 'NE single 4', 'NE single 18'])
    plt.ylabel(metric)
    plt.gcf().subplots_adjust(bottom=0.15)
    plt.gcf().subplots_adjust(left=0.3)

    plt.show()
    fig.savefig(f'../../results/model_performance_{metric.lower().replace(" ", "_")}.png', dpi=300)
```
